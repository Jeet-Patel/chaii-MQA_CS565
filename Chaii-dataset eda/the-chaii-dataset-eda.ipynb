# Analysis on `df_train`

It has a `question` and a `context` (the inputs) and an `answer_text` (the output) plus the `answer_start` position indicator.

Note that the submission only requires the `PredictionString` and not the position of it.

```python
df_train.head()
```

Note: The answer is always a substring of the context:

```python
for _, row in df_train.iterrows():
    assert row.answer_text in row.context
```
